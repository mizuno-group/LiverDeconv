##### Baseline methods with LM13
1. FARDEEP
2. EPIC
3. CIBERSORTx
4. DCQ

```python
import pprint
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

Base_dir = 'C:/github/LiverDeconv' # cloning repository
import sys
sys.path.append(Base_dir)
from Evaluation import evaluator
```

***
##### 1. FARDEEP

```python
res_df = pd.read_csv(Base_dir + '/Evaluation/baseline/FARDEEP_LM13_Results.csv',index_col=0)
val_df = pd.read_csv(Base_dir + '/Data/processed/facs_true_population.csv',index_col=0)

s = val_df.sum()
norm_val = val_df/s

# eval
ev = evaluator.Evaluator()
ev.set_deconv_res(res_df=res_df,z_score=True)
ev.remove_samples(remove_list=["CIV_7","CIV_8","CIP_7","CIP_8"])
ev.set_validation_ref(val_df=norm_val)
ev.process_validation_ref(z_score=True)

# eval
ev = evaluator.Evaluator()
ev.set_deconv_res(res_df=res_df,z_score=True)
ev.remove_samples(remove_list=["CIV_7","CIV_8","CIP_7","CIP_8"])
ev.set_validation_ref(val_df=norm_val)
ev.process_validation_ref(z_score=True)

ev.evaluate(dec_names=[["Neutrophil"],["Monocyte"],["NK"],["Eosinophil"]],
            val_names=[["Neutrophil"],["Monocyte"],["NK"],["Eosinophil"]],sort_index=["APAP","MDA","ANIT","TAA","ConA","GAL","CCl4","Ctrl"],title="Neutrophils",do_plot=True,simple=False,eval_all=False,dpi=100)
pprint.pprint(ev.cor_res)
```
